```python
heart_data = data[data['exists'] == True]
heart_data
#drops all False data
```

Check for null values

```python
heart_data.isna().sum()

#pairwise exclusion: occurs when the statistical procedure uses cases that contain some missing data (https://forum.jamovi.org/viewtopic.php?t=2711#:~:text=Re%3A%20Exclude%20cases%20pairwise%20in%20linear%20regression,-Post%20by%20Abbot&text=Pairwise%20deletion%20occurs%20when%20the,variables%20with%20non%2Dmissing%20values.)
```

Impute values

```python
for col in heart_data:
    if heart_data[col].dtype == 'object':
        heart_data[col].fillna('Not known', inplace=True)
    if heart_data[col].dtype == 'int64':
        heart_data[col].fillna(heart_data[col].mode()[0],inplace=True)
    if heart_data[col].dtype == 'float64':
        heart_data[col].fillna(heart_data[col].mean(),inplace=True)

heart_data
```

```python
heart_data.drop(['id', 'exists'], axis=1, inplace=True)
heart_data

#Drop ID and Exist data
```

```python
heart_data.info()
```

```python
for col in heart_data:
    print(col, heart_data[col].dtype)
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()

for col in heart_data:
    if heart_data[col].dtype == 'object':
        heart_data[col] = heart_data[col].astype(str)
        heart_data[col] = enc.fit_transform(heart_data[col].values.reshape(-1,1))

heart_data
```

```python
from sklearn.cluster import KMeans

#initialize kmeans parameters
kmeans_kwargs = {
"init": "random",
"n_init": 10,
"random_state": 42,
}

#create list to hold SSE values for each k
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(heart_data)
    sse.append(kmeans.inertia_)

#SSE means Sum of Squared Errors - Difference between observed and predicted values.

#visualize results
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()

#plotting elbow method

#The Elbow itself is on the X axis. 3 through 4
```

KMeans

```python
#instantiate the k-means class, using optimal number of clusters
kmeans = KMeans(init="random", n_clusters=6, n_init=10, random_state=42) #edits clusters for everything below (original value is three)

#fit k-means algorithm to data
kmeans.fit(heart_data)

#view cluster assignments for each observation
predictions = kmeans.labels_
predictions
```

**Visualize** clusters

```python
from sklearn.decomposition import PCA

pca = PCA(2) #principal component analysis

pca_data = pca.fit_transform(heart_data)

x = pca_data[:,0]
y = pca_data[:,1]

plt.scatter(x, y, c=predictions, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
print(centers.shape)
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

Prepare submission

```python
submission['cluster'] = predictions
submission.to_csv('submission.csv', index=False)
submission = pd.read_csv('submission.csv')
submission
```
